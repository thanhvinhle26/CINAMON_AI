```python
# Function use for finding the area of tampered image
def count_255(mask):
    i=0
    for row in range(mask.shape[0]):
        for col in range(mask.shape[1]):
            if mask[row,col]==255:
                i+=1
    return i
```

```python
#Generate fake data by finding tampered area with block 64*64
def sample_fake(img, mask):
    kernel_size=64
    stride=32
    
    samples=[]
    
    for y_start in range(0, img.shape[0]-kernel_size+1, stride):
        for x_start in range(0, img.shape[1]-kernel_size+1, stride):
            
            c_255=count_255(mask[y_start:y_start+kernel_size, x_start:x_start+kernel_size])
            
            if (c_255>1600) and (kernel_size*kernel_size-c_255>1600):
                samples.append(img[y_start:y_start+kernel_size, x_start:x_start+kernel_size, :3])    
        
    return samples
```

```python
samples_fake=[]
i=0
for fake, mask in zip(tp_images, mask_images):
    image_samples=sample_fake(fake, mask)
    for sample in image_samples:
        samples_fake.append(sample)
        i+=1
```

```python
#Generate authentic data by dividing in 64*64 blocks
def sample_random(img, num_samples, stride=32):
    kernel_size=64
    
    x_start=0
    y_start=0
    samples=[]
    
    for y_start in range(0, img.shape[0] - kernel_size + 1, stride):
        for x_start in range(0, img.shape[1] - kernel_size + 1, stride):

            #c_255 = count_255(mask[y_start:y_start + kernel_size, x_start:x_start + kernel_size])

            #if (c_255 > threshold) and (kernel_size * kernel_size - c_255 > threshold):
            samples.append(img[y_start:y_start + kernel_size, x_start:x_start + kernel_size, :3])

    
    indices=np.random.randint(0, len(samples), min(len(samples), num_samples))
    
    sampled=[]
    for i in indices:
        sampled.append(samples[i])
    
    return sampled
```

```python
#Function for finding ratio of authentic blocks/ tampered blocks 
def ratio(input):
    return np.sum(input==0,dtype=np.float32)/(np.sum(input==1,dtype=np.float32)+np.sum(input==0,dtype=np.float32))
```

```python
# Predict image is authentic or not
def predict_forensic_image(image):
    input_sample=sample_input(image)
    input_sample=np.array(input_sample)
    prediction=model_aug.predict_classes(input_sample)
    rat=ratio(prediction)
    rat=np.array(rat).reshape(-1,1)
    predict=clf.predict(rat)
    if predict==0:
        print("Authentic image")
    else:
        print("Tampered image")
    plt.imshow(image)
```
